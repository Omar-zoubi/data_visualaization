```python
import pandas as pd 
import matplotlib as plt
import seaborn as sns
```

```python

df = pd.read_csv('./fundamentals.csv')
df.tail(5)
```

# Calculate the avag of Capital Expenditures:

```python

avg_capital_exp = df.groupby("Period Ending",as_index = False).mean()
avg_capital_exp.head(5)
period = df[["Period Ending","Capital Expenditures"]].groupby("Period Ending",as_index = False).mean()
period.head(5)
```

# Calculating avg Total Revenue per a spacific period:

```python
avg_total_rev = df.groupby("Period Ending",as_index = False).mean()
avg_total_rev.head(5)
period = df[["Period Ending","Total Revenue"]].groupby("Period Ending",as_index = False).mean()
period.head(5)
```
